```python
import cv2
import pickle
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from transformers import CLIPProcessor, CLIPModel
from PIL import Image, ImageDraw
from google.colab import files
import os
import requests
from io import BytesIO
from IPython.display import display, Image as IPImage

# Step 1: Load Pre-trained Models
object_detection_model = cv2.dnn.readNet("/content/drive/MyDrive/yolov3.weights", "/content/drive/MyDrive/yolov3.cfg")  # Replace with your model paths
clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")

# Step 2: Load Dataset
with open("/content/drive/MyDrive/products_with_reviews_image_embeddings.pkl", "rb") as f:
    dataset = pickle.load(f)  # Assumes the pickle contains embeddings and image paths
    dataset_embeddings = dataset['embeddings']  # NxD numpy array
    image_paths = dataset['metadata']  # List of metadata objects

# Step 3: Detect and Extract Product
def detect_and_extract(image_path, text_query):
    image = cv2.imread(image_path)
    height, width = image.shape[:2]

    # Object detection using YOLO (Example with COCO classes)
    object_detection_model.setInput(cv2.dnn.blobFromImage(image, 1 / 255.0, (416, 416), swapRB=True, crop=False))
    layer_names = object_detection_model.getLayerNames()
    output_layers = [layer_names[i - 1] for i in object_detection_model.getUnconnectedOutLayers()]
    detections = object_detection_model.forward(output_layers)

    # Load COCO class names
    with open("/content/drive/MyDrive/coco.names", "r") as f:
        classes = [line.strip() for line in f.readlines()]

    # Find bounding box for the text query class
    for output in detections:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5 and classes[class_id] == text_query:
                center_x, center_y, w, h = (detection[0:4] * np.array([width, height, width, height])).astype("int")
                x = int(center_x - w / 2)
                y = int(center_y - h / 2)
                product_frame = image[y:y + h, x:x + w]
                return product_frame

    raise ValueError(f"No product found for the query: {text_query}")

# Step 4: Extract Embedding for the Product Frame
def get_clip_embedding(image):
    pil_image = Image.fromarray(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    inputs = clip_processor(images=pil_image, return_tensors="pt", do_center_crop=True)
    outputs = clip_model.get_image_features(**inputs)
    return outputs.detach().numpy().flatten()

# Step 5: Find Top-5 Similar Images
def find_similar_images(product_embedding, dataset_embeddings, image_paths):
    similarities = cosine_similarity([product_embedding], dataset_embeddings)[0]
    top_indices = similarities.argsort()[-5:][::-1]  # Indices of top-5 similar images
    return [image_paths[i] for i in top_indices]

# Step 6: Display Images
def display_image_from_url(image_url, width):
    response = requests.get(image_url)
    img = Image.open(BytesIO(response.content))
    img = img.resize((width, int(img.height * (width / img.width))))
    draw = ImageDraw.Draw(img)
    display(img)

def display_image(image_path, width):
    img = Image.open(image_path)
    img = img.resize((width, int(img.height * (width / img.width))))
    draw = ImageDraw.Draw(img)
    display(img)

def display_extracted_frame(image_array, width):
    img = Image.fromarray(cv2.cvtColor(image_array, cv2.COLOR_BGR2RGB))
    img = img.resize((width, int(img.height * (width / img.width))))
    draw = ImageDraw.Draw(img)
    display(img)

# Step 7: Main Program
if __name__ == "__main__":
    print("Please upload an image file.")
    uploaded = files.upload()

    if uploaded:
        user_image_path = list(uploaded.keys())[0]
        text_query = input("Enter the product to search for (e.g., 'sofa', 'chair'): ")

        try:
            # Display query image
            print("Query Image:")
            display_image(user_image_path, width=300)

            # Extract product frame
            product_frame = detect_and_extract(user_image_path, text_query)

            # Display extracted product frame
            print("Extracted Product Frame:")
            display_extracted_frame(product_frame, width=300)

            # Get product embedding
            product_embedding = get_clip_embedding(product_frame)

            # Find top-5 similar images
            similar_images = find_similar_images(product_embedding, dataset_embeddings, image_paths)

            print("Top-5 similar images:\n")
            for metadata in similar_images:
                display_image_from_url(metadata['image'], 300)
                print("Title:", metadata['title'])
                print("Brand:", metadata['brand'])
                print("Description:", metadata['description'])
                print("Reviews:", metadata['reviews'][:2])  # Display first 2 reviews
                print("\n---\n")
        except ValueError as e:
            print(e)

        # Remove the uploaded file to clean up
        os.remove(user_image_path)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import precision_recall_fscore_support
from transformers import CLIPProcessor, CLIPModel
from PIL import Image
import torch
from google.colab import files
import requests
from io import BytesIO
import pickle
from IPython.display import display
import cv2
import numpy as np
import time

# Load the CLIP model and processor
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
model.eval()

# Load precomputed embeddings and metadata
pickle_file_path = "/content/drive/My Drive/image_embeddings_amazon_fashion_main.pkl"
with open(pickle_file_path, "rb") as f:
    data = pickle.load(f)

stored_embeddings = data["embeddings"]
stored_metadata = data["metadata"]

# Helper function to fetch image from URL
def fetch_image_from_url(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return Image.open(BytesIO(response.content)).convert("RGB")
    except Exception as e:
        print(f"Error fetching image from URL {url}: {e}")
        return None

# Helper function to display images with a custom width
def display_image_with_width(image, width):
    aspect_ratio = image.height / image.width
    new_height = int(width * aspect_ratio)
    resized_image = image.resize((width, new_height))
    display(resized_image)

# Step 1: Upload the query video
query_video_path = "/content/adidas-tshirt-video.mp4"

# Input the accompanying text query
query_text = input("Enter text query: ")

# Detect and extract a frame with the person's image from the video
def extract_person_frame(video_path):
    cap = cv2.VideoCapture(video_path)
    detected_frame = None

    # Use a pre-trained Haar Cascade for person detection
    person_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_fullbody.xml')

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Convert frame to grayscale for detection
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        persons = person_cascade.detectMultiScale(gray_frame, scaleFactor=1.1, minNeighbors=3, minSize=(50, 50))

        if len(persons) > 0:
            # Return the entire frame where the person is detected
            detected_frame = frame
            break

    cap.release()
    return detected_frame

# Extract the frame with a person
detected_frame = extract_person_frame(query_video_path)

if detected_frame is not None:
    # Convert to RGB and display the image
    query_image = Image.fromarray(cv2.cvtColor(detected_frame, cv2.COLOR_BGR2RGB))
    print("\nExtracted Frame with Person:")
    display_image_with_width(query_image, width=350)
else:
    print("No person detected in the video.")
    exit()

# Step 2: Combine image and text query for similarity
def compute_combined_query_embedding(image, text):
    inputs = processor(text=text, images=image, return_tensors="pt", padding=True)
    with torch.no_grad():
        image_embeds = model.get_image_features(pixel_values=inputs["pixel_values"]).squeeze()
        text_embeds = model.get_text_features(input_ids=inputs["input_ids"]).squeeze()
        # Combine image and text embeddings with more weightage on text
        combined_embeds = (0.3 * image_embeds + 0.7 * text_embeds)  # Weighted average
    return combined_embeds.cpu().numpy()

start_combination = time.time()
combined_query_embedding = compute_combined_query_embedding(query_image, query_text)
combination_time = time.time() - start_combination

# Compute cosine similarity with stored embeddings
start_similarity = time.time()
similarities = cosine_similarity([combined_query_embedding], stored_embeddings)[0]
similarity_time = time.time() - start_similarity

# Sort indices by similarity
sorted_indices = similarities.argsort()[::-1]  # Descending order

# Get the top 5 similar images with the predicted brand
top_indices = sorted_indices[:5]

def evaluate_metrics(top_similarities, ground_truth, threshold=0.5):
    # Binary relevance: 1 if similarity >= threshold, else 0
    predictions = [1 if sim >= threshold else 0 for sim in top_similarities]
    precision, recall, f1, _ = precision_recall_fscore_support(ground_truth, predictions, average="binary")
    return precision, recall, f1

# Step 3: Display the top 5 similar images and calculate metrics
print("\nTop 5 Similar Images with the Predicted Brand and Matching Query:")

ground_truth = [1, 0, 1, 0, 1]  # Example ground truth; modify based on your dataset
for i, index in enumerate(top_indices):
    meta = stored_metadata[index]
    print(f"\nMatch {i+1}:")
    print(f"  Name: {meta['title']}")
    print(f"  Brand: {meta['brand']}")
    image_url = meta['image']
    similar_image = fetch_image_from_url(image_url)
    if similar_image:
        display_image_with_width(similar_image, width=300)
    print(f"  Similarity Score: {similarities[index]:.4f}")

precision, recall, f1 = evaluate_metrics(similarities[top_indices], ground_truth, threshold=0.5)

# Display performance metrics
print("\nPerformance Metrics:")
print(f"Precision: {precision:.2f}, Recall: {recall:.2f}, F1-Score: {f1:.2f}")
print(f"Combination Time: {combination_time:.2f} seconds")
print(f"Similarity Computation Time: {similarity_time:.2f} seconds")
```
